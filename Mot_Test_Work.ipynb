```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dvsa-mot-06-mot-test-stations-and-testers.csv',encoding='latin-1')
```

```python
plt.bar(df['ï»¿Financial year'],df['MOT nominated testers'],color='orange',alpha=1)
plt.ylabel('MOT nominated testers')
plt.xlabel('Financial year')
plt.title('Annual MOT Nominated Testers')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df['ï»¿Financial year'],df['Private MOT test stations'],color='orange',alpha=1)
plt.xlabel('Financial year')
plt.ylabel('Private MOT test stations')
plt.title('Annual Private MOT Test Stations')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df['ï»¿Financial year'],df['MOT nominated testers'],color='deepskyblue',alpha=1,marker='o')
plt.ylabel('MOT nominated testers')
plt.xlabel('Financial year')
plt.title('Annual MOT Nominated Testers')
plt.xticks(rotation=45)
plt.show()
```

These bar graphs are showing the number of MOT nominated testers increasing throughout each financial year. The bar graph showing the testers has been developed annually.

The plot graph showing the number of MOT Nominated Testers increased throughout each financial year. This graph has straight line and it is easy to visualise the data.
