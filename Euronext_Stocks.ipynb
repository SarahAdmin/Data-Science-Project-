```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Euronext Invoice 1.csv',encoding='latin-1')
```

```python
df.columns
```

```python
df.set_index('Description')
```

```python
df.plot()
```

```python
df.plot(kind='barh')
```

```python
plt.bar(df['Description'],df['Apr-23'],color='red',alpha=1)
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Euronext Invoice April 2023')
plt.xticks(rotation=45,ha='right')
plt.axvline(color='black',linestyle='-')
plt.axhline(color='black',linestyle='-')
plt.show()
```

```python
plt.bar(df['Description'],df['Jan-24'],color='red',alpha=1)
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Euronext Invoice January 2024')
plt.xticks(rotation=45,ha='right')
plt.axvline(color='black',linestyle='-')
plt.axhline(color='black',linestyle='-')
plt.show()
```

```python
plt.bar(df['Description'],df['Jan-25'],color='grey',alpha=1)
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Euronext Invoice January 2025')
plt.xticks(rotation=45,ha='right')
plt.axvline(color='black',linestyle='-')
plt.axhline(color='black',linestyle='-')
plt.show()
```

```python
plt.bar(df['Description'],df['Aug-25'],color='grey',alpha=1)
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Euronext Invoice August 2025')
plt.xticks(rotation=45,ha='right')
plt.axvline(color='black',linestyle='-')
plt.axhline(color='black',linestyle='-')
plt.show()
```

```python
plt.plot(df['Description'],df['Aug-25'],color='red',alpha=1,marker='o')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Euronext Invoice August 2025')
plt.xticks(rotation=45,ha='right')
plt.show()
```

The most popular amount of this graph of Euronext is August 2025 because the stocks have investment growth our amount as increased the amount for August 2025 stocks.
