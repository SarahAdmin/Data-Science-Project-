```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HMCTS_MI_output_accessible_Dec_25.csv',encoding='latin-1')
```

```python
plt.bar(df['Month'],df['Civil Claims received'],color='navy',alpha=1)
plt.xlabel('Month')
plt.ylabel('Civil Claims Received')
plt.title('Civil Claims')
plt.xticks(rotation=45)
plt.show()
```

The graph shows the civil claims recieved throughout December 2024 to December 2025. Two most popular months were May 2025 and July 2025 because two months has highest number of claims. The least popular month was December 2024 because least number of claims has been recieved.
