```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car_data.csv',encoding='latin-1')
```

```python
df
```

```python
df.groupby('Make')['Model'].value_counts()
```

```python
plt.barh(df['Price'],df['Model'],color='grey',alpha=1)
plt.xlabel('Price')
plt.ylabel('Model')
plt.title('Price per Model')
plt.axhline(color='navy',linestyle='-',alpha=0.5)
plt.show()
```

```python
plt.plot(df['Model'],df['Price'],color='navy',marker='o',alpha=0.5)
plt.xlabel('Model')
plt.ylabel('Price')
plt.title('Price per Model')
plt.axhline(color='rebeccapurple',linestyle='-',alpha=0.5)
plt.axvline(color='rebeccapurple',linestyle='-',alpha=0.5)
plt.show()
```

```python
df.groupby('Model')['Colour'].value_counts()
```

```python
plt.bar(df['Make'],df['Price'],color='grey',alpha=1)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Price per Make')
plt.show()
```
