```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {'Colour':['Red','Yellow','Green','Blue','Pink','Orange','Purple','White','Black'],'Number_of_Students':[20,18,15,10,11,9,5,7,11]}
final_df = pd.DataFrame(data,index=[1,2,3,4,5,6,7,8,9])
```

```python
final_df
```

```python
plt.pie(final_df['Number_of_Students'],labels=final_df['Colour'])
plt.title('Favourite Colours')
plt.legend()
plt.show()
```

```python
plt.bar(final_df['Colour'],final_df['Number_of_Students'])
plt.xlabel('Colour')
plt.ylabel('No_of_People')
plt.title('Favourite Colour')
plt.show()
```

These charts (Pie and Bar) shows the people have each favourite colours. For example: Lee Cox's favourite colour is Purple. The most popular colour is Red as twenty people have Red as their favourite colour. The least popular colour is Purple as five people have Purple as their favourite colour.
