```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('student1.json')
```

```python
course_count = df["'coursecode'"].value_counts()
```

```python
course_count
```

```python
course_count.plot(kind='bar',title='Courses at Kings University')
```

The bar graph shows the number of people chose the course at Kings University. The most popular course was Film Studies as FST because eight students chose "Film Studies" at university. The least number of students chose "Chemical Engineering" because only student chose this course at university. Other courses on the bar graphs has four same number of students because four students picked Cyber Security and Computer Games Programming courses, six people chose Pharmacy and Early Years courses, eight people chose Sports Science and Law and ten people chose Business and Media Studies course. This graph became even number of students chose two different courses.

```python
course_count.plot(kind='pie',title='Courses at Kings University',legend=True)
```
