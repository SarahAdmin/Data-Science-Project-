```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ticket_data.csv',encoding='latin-1')
```

```python
ticket_count = df['Ticket price'].value_counts()
```

```python
ticket_count
```

```python
plt.pie(ticket_count,labels=ticket_count)
plt.title('Number of Ticket Price purchases')
plt.legend()
plt.show()
```

The Pie chart shows the number of attendees purchased the ticket prices are £0 and £20. 82 attendees paid tickets for £20 and 21 attendees paid tickets for free of charge.

```python
df1 = pd.read_json('StudentPizzaOrder.json')
```

```python
plt.pie(df1['Price'],labels=df1['Topping'])
plt.title('Pizza Topping Orders')
plt.show()
```

```python
final_toppings = df1['Topping'].value_counts()
```

```python
final_toppings
```

```python
plt.pie(final_toppings,labels=final_toppings)
plt.title('Pizza Toppings Orders')
plt.legend()
plt.show()
```

```python
plt.plot(final_toppings,marker='o')
plt.show()
```

On these charts (Pie and plot) are showing the pizza order data. The most popular topping is Meat Feast because students ordered Meat Feast pizza. The least popular toppings are BBQ Chicken and Cheese & Tomato but there are the same.
