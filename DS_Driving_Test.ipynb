```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drt121e-car-driving-tests-with-zero-faults.csv',encoding='latin-1')
```

```python
df.columns
```

```python
df.set_index('Year ')
```

```python
plt.bar(df['Year '],df[' Total car driving tests passed '],color='skyblue',alpha=1,)
plt.xlabel('Year')
plt.ylabel('Total of Car Driving Test Passed')
plt.title('Drivers Passed Car Driving Test')
plt.xticks(rotation=45, ha='right')
plt.axvline(color='navy',alpha=1)
plt.axhline(color='navy',alpha=1)
plt.show()
```

```python
plt.plot(df['Year '],df[' Total car driving tests passed '],color='navy',alpha=1,marker='o')
plt.xlabel('Year')
plt.ylabel('Total of Car Driving Test Passed')
plt.title('Drivers Passed Car Driving Test')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df.rename(columns={'Car driving test pass rate':'Percentage of Car driving test passed'})
```
